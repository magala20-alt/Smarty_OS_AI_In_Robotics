```python
from ultralytics import YOLO
```

```python
import numpy
print(numpy.__version__)
```

```python
# Load pretrained YOLOv8 model
model = YOLO("yolov8m.pt")
```

```python
# Train (transfer learning)
model.train(
    data="datasets/data.yaml",  # dataset config
    epochs=10,
    batch=8,
    imgsz=640,
    lr0=0.01,
    optimizer="SGD",
    momentum=0.9,
    freeze=20,         # freeze backbone layers
    device="cpu",          # cpu
    project="runs/train",  # where results are saved
    name="toy_animals_head",
    verbose=True
)
```

```python
model = YOLO("runs/train/toy_animals_head/weights/best.pt")
model.train(
    data="datasets/data.yaml",
    epochs=30,                 # fine-tuning phase
    batch=4,
    imgsz=640,
    lr0=0.001,                 # smaller LR when unfreezing
    optimizer="SGD",
    momentum=0.9,
    weight_decay=5e-4,
    freeze=0,                  # unfreeze all
    device="cpu",
    project="runs/train",
    name="toy_animals_full",
    verbose=True
)
```

```python
from ultralytics import YOLO

# Load the last or best checkpoint from previous run
model = YOLO("runs/train/toy_animals_full/weights/last.pt")

# Continue training
model.train(
    data="datasets/data.yaml",
    epochs=20,
    batch=4,
    imgsz=640,
    lr0=0.001,
    optimizer="SGD",
    momentum=0.9,
    weight_decay=5e-4,
    device="cpu",
    project="runs/train",
    name="toy_animals_full_resume",
    resume=True,    # set resume to true
    patience=10

)
```
